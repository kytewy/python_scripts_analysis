```python
#!/usr/bin/python3
#
# Assignment 2 - Group 1
# Author: Xi Chen
# Date: Apr 04, 2019
# 

import pandas as pd
import csv

# 1. Column headers to read csv file
dataCol = ['GEO.display-label','HC01_EST_VC01']

# 2. Open/create csv file to write
with open('C:/Group 1 - CSV for Staging/Household_Est.csv', 'w', newline='') as csvfile:
    csvWriter = csv.writer(csvfile)

    # 2.1 Write header row (with columns' name)
    columnNames = ['STATE','HOUSEHOLD','YEAR']

    csvWriter.writerow([columnNames[0],
                         columnNames[1],
                         columnNames[2]])

    # 3. Run a for loop to go though all data file of years from 2005 to 2017
    for year in range(2005, 2018):
        # check to get odd year (original dataset just have odd year)
        if year%2 > 0:
            # 4. Read csv file and load conten to dataframe, just take the column we need
            
            suffix_name = '_1YR_S1901_with_ann.csv'
            
            # file name for 2005 has a different name convention
            if year == 2005:
                suffix_name = '_EST_S1901_with_ann.csv'
                
            data = pd.read_csv('C:/Group 1 - Data Lake/6-Household EST/ACS_' + format((year-2000), '02d') + suffix_name)
            df = pd.DataFrame(data, columns = dataCol)

            # 5. Run a for loop for data item in the dataframe
            for index, row in df.iterrows():
                # Ignore United States data, just get States data
                if row[dataCol[0]] != 'United States':
             
                    print([row[dataCol[0]], row[dataCol[1]], str(year)])

                    # 7. Write data to CSV file
                    csvWriter.writerow([row[dataCol[0]], row[dataCol[1]], year])
```
