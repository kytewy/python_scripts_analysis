```python
#!/usr/bin/python3
#
# Assignment 2 - Group 1
# Author: Hung Nguyen
# Date: Mar 23, 2019
# 

import pandas as pd

import csv

# 1. Column headers to read
dataCol = ['Geography', 'Estimate; SEX AND AGE - 5 to 9 years', 'Estimate; SEX AND AGE - 10 to 14 years', 'Estimate; SEX AND AGE - 15 to 19 years', 'Estimate; SEX AND AGE - 20 to 24 years']
    
# 2. Open/create csv file to write
with open('C:/Group 1 - CSV for Staging/US_State_Under_19.csv', 'w', newline='') as csvfile:
    csvWriter = csv.writer(csvfile)

    # 2.1 Write header row (with columns' name)
    columnNames = ['STATE','FROM_5_TO_9','FROM_10_TO_14','FROM_15_TO_19','FROM_20_TO_24','YEAR']

    csvWriter.writerow([columnNames[0],
                         columnNames[1],
                         columnNames[2],
                         columnNames[3],
                         columnNames[4],
                         columnNames[5]])
    
    # 3. Run a loop for data by years from 2010 to 2018
    for year in range(10, 18):
        
        # 4. Read csv file per each year
        data = pd.read_csv("C:/Group 1 - Data Lake/5-Age Data 2010-2017/ACS_" + str(year) + "_5YR_DP05_with_ann.csv", skiprows=1)

        # 5. Load data to dataframe
        df = pd.DataFrame(data, columns = dataCol)

        # 6. Run a for loop for data in the dataframe
        for index, row in df.iterrows():
            print([row[dataCol[0]], row[dataCol[1]], row[dataCol[2]], row[dataCol[3]], row[dataCol[4]], "20"+str(year)])
            
            # 7. Write data to CSV file
            csvWriter.writerow([row[dataCol[0]], row[dataCol[1]], row[dataCol[2]], row[dataCol[3]], row[dataCol[4]], "20"+str(year)])
```
