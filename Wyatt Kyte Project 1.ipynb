```python
# -*- coding: utf-8 -*-
#Created on Sat Nov 24 21:22:34 2018
#@author: wyatt
# Project 1:  Text Analysis

import tkinter as tk
from tkinter.constants import *

from collections import Counter
import re

# Create the application base class
#=============================
class MainWindow():
    
    # Initialize application (attributes, widgets)
    def __init__(self, master):
        master.title('Project 1')
        
        # Add top label
        #=============================
        
        self.lbTop = tk.Label(master, text="Please Input the file name", width = 33)
        self.lbTop.grid(row = 0, column = 0, columnspan = 2, sticky=W+E)
        
        # Add language buttons & set command
        #=============================
        
        self.entry = tk.Entry(master, width=32)
        self.entry.grid(row = 1, column = 0, padx=5, pady=5)
        
        self.btnSave = tk.Button(master, text="Save output", command = self.doSaveFile)
        self.btnSave.grid(row = 2, column = 1, sticky=W+E)
        
        self.btnExit = tk.Button(master, text="Exit", command = self.doQuit)
        self.btnExit.grid(row = 3, column = 1, sticky=W+E)
        
        self.lbOutput = tk.Label(master, text="-", width = 33, wraplength = 60)
        self.lbOutput.grid(row = 4, column = 0, columnspan = 2, sticky=W+E)
        
    #Methods to be used    
                
    # https://stackoverflow.com/questions/42282840/count-words-in-a-string-without-using-split
    #=============================
    
    def doQuit(self):
         root.destroy()
    
    def doSaveFile(self):
        # 1. get the file name from Entry
        fileName = str(self.entry.get())
        print(fileName)
        
        # 2. open the file in read mode
        with open(r"" + fileName + ".txt") as textFile:
            allText = textFile.read()
        
        # calculate total word count
        totalBlank = allText.count(" ")
        totalCharacter = len(allText) 
        totalNonBlankCharacter = totalCharacter - totalBlank
        totalWord = self.word_count(allText)
        
        # 3. processing the file
        wordList = re.sub("[^\w]", " ",  allText).split()
        print("word count: " + str(len(wordList)))
        
        counts = dict()
        for word in wordList:
            if word in counts:
                counts[word] += 1
            else:
                counts[word] = 1
                
        for key in counts.keys(): 
            print(key + " - count: " + str(counts[key]))
            
        # 4. save the output to txt file
        outputFileName = fileName.replace("clean", "analysis")
        with open(outputFileName + ".txt", "w") as outputFile:
            outputFile.write("Name of text:" + fileName.replace("_clean", "") + "\n")
            outputFile.write("Total Non-blank Character Count:" + str(totalNonBlankCharacter) + "\n")
            outputFile.write("Total Blank Character Count:" + str(totalBlank) + "\n")
            outputFile.write("Percentage Blank Character:" + str(totalBlank/totalCharacter*100) + "%\n")
            outputFile.write("Total Word Count:" + str(totalWord) + "\n")
            outputFile.write("Word, Count:\n")
            for key in sorted(counts, key=counts.get, reverse=True):#dict(sorted(counts.items())):#counts.keys():
                outputFile.write(key + "," + str(counts[key]) +"\n")
            print("File saved!")   
    
    
    def is_word_character(self,c):
        return 'a' <= c <= 'z' or 'A' <= c <= 'Z' or '0' <= c <= '9' or c == '_'

    def word_count(self,str):
        c = 0
        for i in range(1, len(str)):
            if not self.is_word_character(str[i]) and self.is_word_character(str[i-1]):
                c += 1
        if self.is_word_character(str[-1]):
            c += 1
        return c
    

# Create the application
#=============================
root = tk.Tk()   
app = MainWindow(root)

# Start the application
#=============================
root.mainloop()
```
