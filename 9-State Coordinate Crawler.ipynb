```python
#!/usr/bin/python3
#
# Assignment 2 - Group 1
# Author: Hung Nguyen
# Date: Mar 21, 2019
# 

import requests
import json

import csv

# 1 Open CSV file to get list of US States
with open('C:/Group 1 - CSV for Staging/Happiness_Score.csv') as csvfile:
    csv_reader = csv.reader(csvfile, delimiter=',')
    headers = next(csv_reader)
    
    # 2 Open CSV file to write coordinate data
    with open('C:/Group 1 - CSV for Staging/State_Coord.csv', 'w', newline='') as csvfile:
        csvWriter = csv.writer(csvfile)

        # 2.1 Write header row (with columns' name)
        columnNames = ['STATE', 'LATITUDE', 'LONGITUDE']  

        csvWriter.writerow([columnNames[0],
                             columnNames[1],
                             columnNames[2]])

        # 2.2 Run for loop for each state
        for row in sorted(csv_reader, key=lambda k: k[1]):
            print('---\n' + row[1])
            
            # 1. Send request to Here map api to get coordinate of each state
            response = requests.get('https://geocoder.api.here.com/6.2/geocode.json?app_id=Z1UQ7pkpsHG7lQWZKcKL&app_code=qhmA-xTUnRtKC_BIRKj9gA&searchtext=' + row[1])
            json_data = json.loads(response.text)
            
            # 2. Print out the lat and long got from JSON Parser
            print(json_data['Response']['View'][0]['Result'][0]['Location']['DisplayPosition']['Latitude'])
            print(json_data['Response']['View'][0]['Result'][0]['Location']['DisplayPosition']['Longitude'])

            # 3. Write data to CSV file
            csvWriter.writerow([row[1],
                                 json_data['Response']['View'][0]['Result'][0]['Location']['DisplayPosition']['Latitude'],
                                 json_data['Response']['View'][0]['Result'][0]['Location']['DisplayPosition']['Longitude']])
```
