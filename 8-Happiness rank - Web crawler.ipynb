```python
#!/usr/bin/python3
#
# Assignment 2 - Group 1
# Author: Hung Nguyen
# Date: Mar 18, 2019
# 

import requests
from bs4 import BeautifulSoup

import csv

# 1. Initialize the request to website to get content
page = requests.get("https://www.businessinsider.com/happiest-us-states-ranked-2018-7")

page.content

# 2. Initialize the BeautifulSoup object to parse HTML content of request
soup = BeautifulSoup(page.content, 'html.parser')

lstStates = soup.find_all('div', class_='slide')

# 3. Open CSV file to get list of US States
with open('C:/Group 1 - CSV for Staging/Happiness_Score.csv', 'w', newline='') as csvfile:
    csvWriter = csv.writer(csvfile)

    # 3.1 Write header row (with columns' name)
    columnNames = ['RANK','STATE', 'FINAL_SCORE', 'HEALTH_RANK', 'LIFESTYLE_RANK', 'ECONOMIC_STABILITY', 'YEAR']

    csvWriter.writerow([columnNames[0],
                         columnNames[1],
                         columnNames[2],
                         columnNames[3],
                         columnNames[4],
                         columnNames[5],
                         columnNames[6]])

    # 3.2 Write data get from list
    for state in lstStates:
        print('---')
        # 1. get state name
        stateName = list(state)[1].find(class_="slide-title-text").get_text()
        
        rank = stateName.split('. ')[0]
        stateName = stateName.split('. ')[1]
        print(stateName)

        # 2. state info
        info = list(state)[1].find('div', class_='')

        #print(info.prettify())
        lst = info.find_all('p')
        
        lstRank = []
            
        for item in lst:
            print(item.get_text())
            lstRank.append(item.get_text().split(': ')[1])
            
        csvWriter.writerow([rank,
                             stateName,
                             lstRank[0],
                             lstRank[1],
                             lstRank[2],
                             lstRank[3],
                            2017])
```
